### 1. Generate random samples from a normal distribution with use of numpy.random.normal module for each set of mean and variance.

```python
# loc is mean
# scale is standard deviation 
# to get the standard deviation, you calculate the square root of the variance
sample_1 = np.random.normal(loc=0,scale=0.2**0.5,size=100)
sample_2 = np.random.normal(loc=0,scale=1.0**0.5,size=100)
sample_3 = np.random.normal(loc=0,scale=5.0**0.5,size=100)
sample_4 = np.random.normal(loc=-2,scale=0.5**0.5,size=100)
```

### 2. Build a python function for a probability density function with use of the formula given above. The python function should return a probability of the random variable falling within a particular range of values.

```python
def pdf(mean, std, x):
    a = 1/(np.sqrt(2*np.pi)*std)
    b = np.exp(-0.5*((x-mean)/std)**2)
    return a*b
```

### 3. Build plot for random samples and probabilities for each normal distribution with different mean variance sets (Hint: x-axis should be random samples and y-axis should be probabilities).

```python
prob1 = pdf(0,0.2**0.5,sample_1)
prob2 = pdf(0,1.0**0.5,sample_2)
prob3 = pdf(0,5.0**0.5,sample_3)
prob4 = pdf(-2,0.5**0.5,sample_4)
```

```python
import seaborn as sns
plot1 = sns.lineplot(sample_1, prob1, color='red')
plot1
```

```python
plot2 = sns.lineplot(sample_2, prob2, color='green')
plot2
```

```python
plot3 = sns.lineplot(sample_3, prob3, color='blue')
plot3
```

### 3. Run independent t-test using stats.ttest_ind to test whether there is significant differences in blood pressures between male group and female group after the treatment. Also, interpret the result of independent t-test between male group’s blood pressures and female group’s blood pressures after the treatment.

#### null hypothosis (Ho): male's blood pressures after treatment = female's blood pressures after treatment
#### alternative hypothosis (Ha): male's blood pressures after treatment != female's blood pressures after treatment
#### alpha = 0.05
#### if alpha>0.05: fail to regect Ho
#### if alpha<0.05: reject Ho

```python
import scipy.stats as stats
male_blood_pressure = blood_pressure[blood_pressure['sex']=='Male']
female_blood_pressure = blood_pressure[blood_pressure['sex']=='Female']
t_statistics, p_value = stats.ttest_ind(male_blood_pressure['bp_after'],female_blood_pressure['bp_after']) 
print(t_statistics,p_value)
```

#### Since the p-value is 0.001 which is less than the threshold which is the alpha(0.05), we reject the null hypothesis. Male's blood pressures after treatment ARE NOT the same as female's blood pressures after treatment. In other words, there is a significant differences in blood pressures between male group and female group after the treatment.

### 4. Often, there are assumptions that need to be satisfied for the test results to be considered valid. Here, we test the normality. This test of normality applies to the differences in values between the groups (Hint: blood pressures after the treatment between the male and female). One method of testing the normality is the Shpiro-Wilk test. This can be completed using the stats.shapiro. Test the normality with use of stats.shapiro and interpret the results (normally distributed or not).

#### null hypothosis (Ho): the difference between male's and women's blood pressures after treatment is normally distributed
#### alternative hypothosis (Ha): the difference between male's and women's blood pressures after treatment is not normally distributed
#### alpha = 0.05
#### if alpha>0.05: fail to regect Ho
#### if alpha<0.05: reject Ho

```python
female_blood_pressure = female_blood_pressure.reset_index(drop=True)
sampling_difference = male_blood_pressure['bp_after']-female_blood_pressure['bp_after']

stats.shapiro(sampling_difference)
```

#### Since the p-value of the difference between male's and women's blood pressures after treatment is 0.7147 which is more than the threshold which is the alpha(0.05), we fail to reject the null hypothesis. We do not have sufficient evidence to say that sample does not come from a normal distribution. In other words, the sample IS normally distributed.

### 5. Another method is visually checking the distribution with a probability plot with or without the correlation value to assess the observed values correlations with the theoretical distribution in question. This can be completed by using stats.probplot and interpret the plot.

```python
fig = plt.figure(figsize= (20, 10))
ax = fig.add_subplot(111)

normality_plot, stat = stats.probplot(sampling_difference, plot= plt,
                                      rvalue= True)
ax.set_title("Probability plot of sampling difference", fontsize= 20)
ax.set

plt.show()
```

#### The points in the plot are all tightly gathered around the line of best fit and are following a relatively straigt line. This validates the previouse finding from question 4 that indicated the sample was normaly distributed. 

### 6. One of the assumptions is that both groups have equal variances. One method for testing this assumption is the Levene’s test of homogeneity of variances. This can be completed using the stats.levene. Run the Levene’s test to test whether the variances from the male group after the treatment and the female group after the treatment are not significantly different. Interpret the results.

#### null hypothosis (Ho): male's blood pressures after treatment and female's blood pressures after treatment have equal variances
#### alternative hypothosis (Ha): male's blood pressures after treatment and female's blood pressures after treatment have unequal variances
#### alpha = 0.05
#### if alpha>0.05: fail to regect Ho
#### if alpha<0.05: reject Ho

```python
Male = male_blood_pressure['bp_after']
Female = female_blood_pressure['bp_after']
stats.levene(Male, Female, center='median')
```

#### Since the p-value is 0.0265 which is less than the threshold which is the alpha(0.05) then we reject the null hypothesis. We have sufficient evidence to say that male's blood pressures after treatment and female's blood pressures after treatment have unequal variances. In other words, the two group's variences are significantly different.
